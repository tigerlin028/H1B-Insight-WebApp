```python
import pandas as pd
```

```python
h1b_df = pd.read_csv("/Users/abcyqqq/Downloads/H1B_data.csv")
h1b_df
```

```python
h1b_df = h1b_df.reset_index(drop = True)
h1b_df
```

```python
h1b_df = h1b_df.drop(columns='bcn')
```

```python
h1b_df.index.name = "h1b_id"
```

```python
h1b_df
```

```python
h1b_df = h1b_df[(h1b_df['gender'] == "male")|(h1b_df['gender'] == "female")]
```

```python
employer_counts = h1b_df['employer_name'].value_counts()
employers_to_keep = employer_counts[employer_counts >= 100].index
h1b_filtered_df = h1b_df[h1b_df['employer_name'].isin(employers_to_keep)]
```

```python
h1b_filtered_df['status_type'].unique()
```

```python
import numpy as np
h1b_filtered_df['status'] = np.where(h1b_filtered_df['status_type'] == 'SELECTED', 1, 0)
```

```python
h1b_filtered_df.reset_index(drop = True, inplace= True)
```

```python
len(h1b_filtered_df['employer_name'].unique())
```

```python
h1b_filtered_df.to_csv("H1B.csv")
```
